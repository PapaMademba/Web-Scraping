```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from itertools import groupby
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
liste_liens = pd.read_csv("Liens_pages.csv")
print(liste_liens)
print(len(liste_liens))
```

### <center> Trouver tous les liens des produits et les exporter

```python
tableau=[],[]
url_lampes = []
for i in range(len(liste_liens)):
    url_lampes.append(liste_liens.iloc[i]["Liens"])
    html = urlopen(url_lampes[i])
    soup = BeautifulSoup(html, 'html')
    type(soup)
    attributs = soup.select("a[class=jsQs]")
    for link in attributs:
        lien=(link.get("href"))
        tableau[0].append(lien)
   
    
    
tab=np.array([tableau[0]])
print(len(tableau[0]))
df=pd.DataFrame(tab, index= ['Liens'])
dft = pd.DataFrame.transpose(df)

print(dft)
export_csv = dft.to_csv(r'Liens_lampadaires.csv', index=None)
                    
```
