```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import re
```

```python
Liens_spots = pd.read_csv("Liens_spots.csv")
print(len(Liens_spots))

Liens_spots= np.unique(Liens_spots)
tab=np.array([Liens_spots])
Liens_spots=pd.DataFrame(tab, index= ['Liens'])
Liens_spots= pd.DataFrame.transpose(Liens_spots)
print(len(Liens_spots))
print(Liens_spots)


```

```python
url_spots=[]
for i in range(len(Liens_spots)):
    url_spots.append(Liens_spots.iloc[i]["Liens"])



html = pd.read_html(url_spots[0])
df = html[0]

df = pd.DataFrame(df).T
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df hea

df = df.loc[:,~df.columns.duplicated()]
print(df)




df3=[]
tabless=[]
for i in range(len(Liens_spots)):    
    try:
        html = pd.read_html(url_spots[i])
        
    except ValueError:
        print(url_spots[i])
        pass
        
    df2=html[0]
    
    df3.append(pd.DataFrame(df2).T)
    new_head = df3[i].iloc[0]
    df3[i] = df3[i][1:]
    df3[i].columns=new_head   
    
    df3[i] = df3[i].loc[:,~df3[i].columns.duplicated()]
    
    for c in df3[i].columns:
        if c not in df.columns:
            df[c] = None
    for c in df.columns:
        if c not in df3[i].columns:
            df3[i][c] = None
df_final=pd.DataFrame.append(df, df3, sort=True)
df_final 


pd.DataFrame.to_excel(df_final, "Spots.xlsx", sheet_name='Sheet1', header=True, index=None)
```

```python
url_spots = []
tableau2 = [],[],[]
for i in range(len(Liens_spots)):
    try:
        url_spots.append(Liens_spots.iloc[i]["Liens"])
        html = urlopen(url_spots[i])
        soup = BeautifulSoup(html, 'html')
        #print(url_lampes)
        nom = soup.select("h1[itemprop=name]")
        str_cells = str(nom)
        cleantext = BeautifulSoup(str_cells, "lxml").get_text().strip()
        prix=soup.select("span[itemprop=price]")
        str_prix = str(prix)
        cleantext_prix = BeautifulSoup(str_prix, "lxml").get_text().strip()
    except AttributeError:
        print("error")
        continue
        
    tableau2[1].append(cleantext)
    tableau2[0].append(url_spots[i])
    tableau2[2].append(cleantext_prix)
    

    
tab=np.array([tableau2[0],tableau2[1], tableau2[2]])
print(len(tableau2))
data=pd.DataFrame(tab, index= ['Liens', 'Nom', "Prix"])
dft = pd.DataFrame.transpose(data)


pd.DataFrame.to_excel(dft, "Spots2.xlsx", sheet_name='Sheet1', header=True, index=None)
```
