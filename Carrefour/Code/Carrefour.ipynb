### <center>Récupération de tous les sites à scraper


```python
tableau=[],[]
url = ["https://www.carrefour.fr/s?q=Chips&noRedirect=1","https://www.carrefour.fr/s?q=Chips&noRedirect=1&page=2"]

for i in range(len(url)):
    time.sleep(120)
    class AppURLopener(urllib.request.FancyURLopener):
        version = "Mozilla/71.0"
    opener= AppURLopener()

    response = opener.open(url[i])
    soup = BeautifulSoup(response, 'html.parser')
     
    attributs = soup.select("a[class='product-card-title product-card-title--food']")
    for link in attributs:
        lien=(link.get("href"))
        lien = "https://www.carrefour.fr" + lien
        tableau[0].append(lien)
   
    
    
tab=np.array([tableau[0]])
print(len(tableau[0]))
df=pd.DataFrame(tab, index= ['Liens'])
dft = pd.DataFrame.transpose(df)

print(dft)
export_csv = dft.to_csv(r'Liens_chips.csv', index=None)
        
```

```python
Liens_chips = pd.read_csv("Liens_chips.csv")

Liens_chips= np.unique(Liens_chips)
tab=np.array([Liens_chips])
Liens_chips=pd.DataFrame(tab, index= ['Liens'])
Liens_chips= pd.DataFrame.transpose(Liens_chips)
print(len(Liens_chips))
print(Liens_chips)
```

```python
liste_url =[]
tableau = [],[],[],[]
link=[],[]


for i in range(len(Liens_chips)):
    
    liste_url.append(Liens_chips.iloc[i]["Liens"])
    link[0].append(Liens_chips.iloc[i]["Liens"])
    link[0][i] = link[0][i][24:len(link[0][i])]
    time.sleep(70)
    class AppURLopener(urllib.request.FancyURLopener):
        version = "Mozilla/71.0"
    opener= AppURLopener()
    
    response = opener.open(urllib.parse.unquote(liste_url[i]))
    soup = BeautifulSoup(response, "html.parser")
    print(i)

    try:
        #nom=soup.select("img[alt=Nutri-score]")["src"]
        nutri = soup.find("img", {"alt":"Nutri-score"})['src']

        str_cells1 = str(nutri)
        nutriscore = BeautifulSoup(str_cells1, "lxml").get_text().strip()
        nutriscore = nutriscore[31]

        nom = soup.find("a", {"href":link[0][i]})
        children = nom.findChildren("span" , recursive=False)
        str_cells2 = str(children)
        nom_clean = BeautifulSoup(str_cells2, "lxml").get_text().strip()

        div = soup.find('div', {'class': 'pdp-card__header--text'})
        poids = div.findChildren("span" , recursive=False)
        str_cells3 = str(poids)
        poids_clean = BeautifulSoup(str_cells3, "lxml").get_text().strip()
    except ValueError:
        print(liste_url[i])
        continue
    except TypeError:
        print(liste_url[i])
        continue
        
    tableau[3].append(nutriscore)
    tableau[0].append(liste_url[i])
    tableau[1].append(nom_clean)
    tableau[2].append(poids_clean)
    
tab=np.array([tableau[0],tableau[1], tableau[2], tableau[3]])
#print(len(tableau))
data=pd.DataFrame(tab, index= ['Liens', 'Nom', "Poids","Nutriscore"])
dft = pd.DataFrame.transpose(data)


pd.DataFrame.to_excel(dft, "attributs_chips.xlsx", sheet_name='Sheet1', header=True, index=None)


```
